## OverConfidence

The Overconfidence is calculated (Adel and Mariem, 2013) by using the following formula:

$$TR = \frac{V_{i,t}}{N_{i,t}}$$

Where:
- \(V_{i,t}\) is the number of traded shares of stock \(i\) (volume traded per day);
- \(N_{i,t}\) represents the number of exchanges of shares \(i\) (number of transactions per day).

This expression has been used as a measure of the volume and number of transactions (Adel and Mariem, 2013).
